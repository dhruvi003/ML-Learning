```python
# we also have country_code dataset
df_country = pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# can we combine these two dataframes?
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
# we are trying to find out how many different coutries are there
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
# we get different country name
# we store it in different variable because we will be plotting pychart in future
```

```python
final_df.Country.value_counts().values # we get records only (not name)
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# now, we will be creating pie chart
# i wish to see, which country has the maximum online order
# we are taking for top 3 countries, otherwise, the pie chart is really jumbled up
# to see percentage, we add 'autopct' parameter
plt.pie(country_values[:3], labels = country_names[:3], autopct='%1.3f%%' )
```

```python
# observation: max record of transactions are from India, US, UK
```

```python
# 2) numerical variables
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
# we can say that, when rating color is white, Aggregate_rating is 0
# rating is red -> 1.8 - 2.4 -- poor
# rating is orange -> 2.5 - 3.4 -- average
#

# now we will converet it in dataframe by _reset_index() and renaming new column
```

```python
ratings
```

```python
# now we wish to find out relationship of ratings with countries
# conclusions
"""
1) rating is from 4.5 to 4.9 -> excellent
2) rating is from 4.0 to 4.4 -> very good
3) rating is from 3.5 to 3.9 -> good
4) rating is from 2.5 to 2.4 -> average
5) rating is from 1.8 to 2.4 -> poor
"""
```

```python
print(ratings.columns.tolist())
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.setdefault('toolbar', 'toolbar2')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,6)

sns.barplot(
    x='Aggregate rating',
    y='Rating Count',
    hue='Rating color',
    data=ratings
)
plt.show()
```

```python
# let's plot count plot
# plotting with respect to categorical value
sns.countplot(x='Rating color', data = ratings, palette='magma')
# this is frequence, not count
```

```python
# find the countries that has given 0 rating
# final_df.head(5)
# country_and_rating = final_df[['Aggregate rating','Country']]
second_df = pd.merge(ratings, final_df[['Aggregate rating','Country']], on='Aggregate rating', how='left')
second_df[second_df['Aggregate rating'] == 0.0]['Country'].value_counts()
```
