```python
import pandas as pd
```

#### Standardization


```python
df = pd.read_csv('/content/train.csv', usecols=['Survived','Pclass','Age','Fare'])
df.head()
```

```python
# replace null values in Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# perform standardization
"""
we use standarscaler library.
"""
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled
```

```python
pd.DataFrame(df_scaled)
```

#### Min Max Scaler


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
min_max_df = pd.DataFrame(min_max.fit_transform(df), columns = df.columns)
min_max_df
```

#### Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
rb_scaler = RobustScaler()
```

```python
rb_scaler_df = rb_scaler.fit_transform(df)
```

```python
pd.DataFrame(rb_scaler_df, columns = df.columns)
```
